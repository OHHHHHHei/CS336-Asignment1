```python
import regex as re

special_tokens = ["<special1>", "<special2>", "<special3>"]

text = "这是一个测试文本，包含<special1> <special2> <special3> 以及一些普通的文本。"

special_regax = "|".join(re.escape(token) for token in special_tokens)

parts = re.split(f"({special_regax})", text)
new = re.split(f"{special_regax}", text)
train_segments = [part for part in parts if part not in special_tokens]
print(parts)
print(new)
print(train_segments)
```

```python
import regex as re

gpt2_pat = re.compile(r"""'(?:[sdmt]|ll|ve|re)| ?\p{L}+| ?\p{N}+| ?[^\s\p{L}\p{N}]+|\s+(?!\S)|\s+""")

text = "Hello, world! This is a test. Let's see how it works."

token = gpt2_pat.findall(text)
print(token)
```
